# Importing the libraries we will use in the project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the dataset using pandas

```python
df = pd.read_csv("Airbnb_Open_Data.csv")
df.head()
```

# Checking the number of rows and columns in the dataset

```python
df.shape
```

# Checking Column Data Types

```python
df.info()
```

# Changing 'last review' column type to date

```python
df['last review'] = pd.to_datetime(df['last review'])
```

```python
df.info()
```

```python
df.describe()
```

# Correlation Between Rating and Price

```python
correlation_coefficient = df['review rate number'].corr(df['price'])
print("Pearson Correlation Coefficient between Rating & Price = ", correlation_coefficient)
```

## We can see that there is little to no correlation between rating and price

# Next we will calculate some KPIs

```python
print("Average listing price = ", df['price'].mean(), "$")
print("Average listing rating = ", df['review rate number'].mean())
print("Average number of reviews per listing = ", df['number of reviews'].mean(), "reviews")
print("Maximum number of reviews per listing = ", df['number of reviews'].max(), "reviews")
```

# Filtering all years of construction per listing

```python
construction_years = df['Construction year'].unique()
print("Construction Years", sorted(construction_years))
```

# Plotting the distribution of number of reviews

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df, x="number of reviews", bins = 50, color='green' )
plt.xlabel("Number of Reviews")
plt.ylabel("Frequency")
plt.title("Distribution of Number of reviews per listing")
plt.show()
```

# We can deduct that the majority of listings have little to no reviews

# Plotting the distribution of review rating

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df, x="review rate number", bins=5, color='green')
plt.xlabel("Review Rating")
plt.ylabel("Frequency")
plt.title("Distribution of review ratings per listing")
plt.show()
```

# We can deduct that fewer number of listings have rating between 1 & 2

# Distribution of Cancellation Policy per listing

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='cancellation_policy')
plt.xlabel("Cancellation Policy")
plt.ylabel("Frequency")
plt.title("Distribution of Cancellation Policy")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='room type', hue='instant_bookable')
plt.xlabel("Room Type")
plt.ylabel("Frequency")
plt.title("Distribution of Room Type per listing")
plt.show()
```

# Distribution of Total Price per Neighbourhood

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df, x="neighbourhood group", y="price", ci=None)
plt.xlabel("Neighbourhood")
plt.ylabel("Price")
plt.title("Distribution of Average Price per Neighbourhood")
plt.show()
```

# Total Listings by Confirmed ID vs Unconfirmed IDs

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x="host_identity_verified")
plt.xlabel("Host ID")
plt.ylabel("Number of Listings")
plt.title("Total Listings by ID Type")
plt.show()
```

# We can deduct that a lot of users have not confirmed their identity on Airbnb
